```
import pandas as pd
```

```
import io
df_surprise = pd.read_csv(io.BytesIO(uploaded['surpriseLabel.csv']))
```

```
surpriseValues = df_surprise.values.tolist()
```

```
surpriseValues
```

```
df_surprise
```

```
d = []
```

```
for i in range(len(surpriseValues)):
  temp = str(surpriseValues[i])
  d.append(str(temp))
```

```
#create format function for every label 
def format(label_name): 
  
    label_name =''.join(label_name)
  
    label_name = str(label_name)
    label_name = label_name.replace(',','')
    label_name = label_name.replace('  ','')

    label_name = label_name.replace("'", "")

    label_name = label_name.split(' ')
   
    return label_name
```

```
d = format(d)
```

```
d
```

```
len(d)
```

```
df_fearLabels = pd.DataFrame(d)
```

```
df_fearLabels = pd.DataFrame({'label': d})
```
