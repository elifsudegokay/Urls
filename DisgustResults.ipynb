```
for i in range(len(disgustValues)):
  temp = str(disgustValues[i])
  d.append(str(temp))
```

```
#create format function for every label 
def format(label_name): 
  
    label_name =''.join(label_name)
  
    label_name = str(label_name)
    label_name = label_name.replace(',','')
    label_name = label_name.replace('  ','')

    label_name = label_name.replace("'", "")

    label_name = label_name.split(' ')
   
    return label_name
```

```
d = format(d)
```
