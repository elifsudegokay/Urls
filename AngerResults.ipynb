```
import pandas as pd
```

```
import io
df_anger = pd.read_csv(io.BytesIO(uploaded['angerLabel.csv']))
```

```
angerValues = df_anger.values.tolist()
```

```
angerValues
```

```
df_anger
```

```
d = []
```

```
for i in range(len(angerValues)):
  temp = str(angerValues[i])
  temp = temp[2:-2]
  d.append(str(temp))
```

```
#create format function for every label 
def format(label_name): 
  
    label_name =''.join(label_name)
  
    label_name = str(label_name)
    label_name = label_name.replace(',','')
    label_name = label_name.replace('  ','')

    label_name = label_name.replace("'", "")

    label_name = label_name.split(' ')
   
    return label_name
```

```
d = format(d)
```

```
d
```

```
len(d)
```

```
df_angerLabels = pd.DataFrame(d)
```

```
df_angerLabels = pd.DataFrame({'label': d})
```

```
from google.colab import files

df_angerLabels.to_csv('df_angerLabels.csv')
files.download('df_angerLabels.csv')
```
